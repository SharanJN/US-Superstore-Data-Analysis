```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('US Superstore.csv', encoding = 'latin-1')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors = 'coerce', dayfirst = True)
```

```python
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
```

```python
regions = df['Region'].unique()
sub_categories = df['Sub-Category'].unique()

for region in regions:
    for sub_category in sub_categories:
        region_category_data = df[(df['Region'] == region) & (df['Sub-Category'] == sub_category)]
        yearly_sales = region_category_data.groupby('Order Year').agg({'Sales': 'sum'}).reset_index()
        plt.figure(figsize=(10, 5))
        plt.bar(yearly_sales['Order Year'], yearly_sales['Sales'], color='skyblue', edgecolor='black')
        plt.xlabel("Year")
        plt.ylabel("Total Sales")
        plt.title(f"Yearly Sales for {sub_category} in {region} Region")
        plt.xticks(yearly_sales['Order Year'])
        plt.show()
```
